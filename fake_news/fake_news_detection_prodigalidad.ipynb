See data text in our datasets

```python
data_text = data.text
data_text
```

```python
import re
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def normalize_text(text):
    # Remove X user handle starting with "@"
    text=re.sub(r'@\w+', '', text)


    # Remove non-alphanumeric characters and extra whitespaces
    text=re.sub(r'[^a-zA-Z\s]', '', text)

    # Remove multiple whitespaces
    text=re.sub(r'\s+', ' ', text)

    # Convert data text to lowercase
    text = text.lower()

    # Apply lemmatization
    text = ' '.join([lemmatizer.lemmatize(w) for w in text.split()])

    return text
```

```python
data_text = data_text.apply(normalize_text)
data_text
```

```python
import nltk
nltk.download('wordnet')

data_target = data.Label
data_target
```

```python
# Transform "R" and "F" to a numeric for easy data modeling
data_target.replace('R', 1, inplace=True)
data_target.replace('F', 0, inplace=True)
data_target
```

```python
tf_idf = TfidfVectorizer()
data_text_transformed = tf_idf.fit_transform(data_text)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    data_text,
    data_target,
    test_size=0.2,
    random_state=42
)

print("x_train Shape:", x_train.shape)
print("x_test Shape:", x_test.shape)
print("y_train Shape:", y_train.shape)
print("y_test Shape:", y_test.shape)
```

SVM Model

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(x_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(x_test)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# Model Precision:
print("Precision:",metrics.precision_score(y_test, y_pred))

# Model Recall:
print("Recall:",metrics.recall_score(y_test, y_pred))

#Model F1:
print("F1 Score:", metrics.f1_score(y_test, y_pred))
```

Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier

# Create a Decision Tree Classifier object
dt_model = DecisionTreeClassifier(random_state=42)

# Train the Decision Tree Model
dt_model.fit(x_train, y_train)

# Predict the response for test dataset
y_pred_dt = dt_model.predict(x_test)

```

```python
print("Accuracy : ",
          accuracy_score(y_test, y_pred_dt)*100)
print("Classification Report of Decision Tree Model")
print("Report : ",
          classification_report(y_test, y_pred_dt))
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Placeholder function for cal_accuracy
def cal_accuracy(y_test, y_pred):
    print("Confusion Matrix: ",
          confusion_matrix(y_test, y_pred))
    print("Accuracy : ",
          accuracy_score(y_test, y_pred)*100)
    print("Report : ",
          classification_report(y_test, y_pred))
```

#### Random Tree Forest Model

```python
random_forest_model = Pipeline([
    ('rf_model', RandomForestClassifier(criterion='gini', n_estimators=400, max_depth=600, random_state=44))
])

random_forest_model.fit(x_train, y_train)
```

Random Forest Model Check

```python
print("Pipeline Model Train Score: ", random_forest_model.score(x_train, y_train))
print("Pipeline Model Test Score: ", random_forest_model.score(x_test, y_test))
```

```python
y_pred = random_forest_model.predict(x_test)
```

```python
df = pd.DataFrame()
df["Actual"], df["Predict"] = y_test, y_pred
df.replace(1, "R", inplace=True)
df.replace(0, "F", inplace=True)
df
```

```python
c_matrix = confusion_matrix(y_test, y_pred)
target_names = ['F', 'R']
sns.heatmap(
    c_matrix,
    annot=True,
    fmt='g',
    center=True,
    cmap='Purples',
    xticklabels=target_names,
    yticklabels=target_names,
)
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix for Random Tree Forest")
plt.show()
c_matrix
```

```python
rf_classification_report = classification_report(y_test, y_pred, target_names=target_names)
print("Classification Report of Random Tree Forest\n", rf_classification_report)
```

#### Logistic Regression

```python
# Find the best parameter for LR
lr_model = LogisticRegression()

param_grid = {
    'C': [0.01, 0.1, 1, 10],
    # 'fit_prior': [True, False],
}

grid = GridSearchCV(lr_model, param_grid, cv=5)
grid.fit(x_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best Cross-Validation Score:", grid.best_score_)
```

```python
lr_model = Pipeline([
    ("lr_model", LogisticRegression(penalty='l1', solver='saga', C=10, random_state=44, max_iter=10000))
])

lr_model.fit(x_train, y_train)
```

Logistic Regression Model Check

```python
print("Pipeline Model Train Score:", lr_model.score(x_train, y_train))
print("Pipeline Model Test Score:", lr_model.score(x_test, y_test))
```

```python
y_pred = lr_model.predict(x_test)

df = pd.DataFrame()
df["Actual"], df["Predict"] = y_test, y_pred
df.replace(1, "R", inplace=True)
df.replace(0, "F", inplace=True)
df
```

```python
c_matrix = confusion_matrix(y_test, y_pred)
target_names = ['F', 'R']
sns.heatmap(
    c_matrix,
    annot=True,
    fmt='g',
    center=True,
    cmap='Purples',
    xticklabels=target_names,
    yticklabels=target_names,
)
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix for Logistic Regression")
plt.show()
c_matrix
```

```python
lr_classification_report = classification_report(y_test, y_pred, target_names=target_names)
print("Classification Report of Logistic Regression\n", lr_classification_report)
```
